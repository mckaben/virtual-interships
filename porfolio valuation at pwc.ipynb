Based on the provided data, compute the historical repayment percentages, i.e. every repaymentâ€™s share of the origination amount

```python
dff = hist_repay_perc.copy()

# Calculate historical repayment percentages
repayment_cols = dff.columns[2:]
for col in repayment_cols:
    dff[col + '_Repay_perc'] = round((dff[col] / dff['Origination Amount']) * 100, 2)

# Display the DataFrame with historical repayment percentages
historical_repayment_percentages = dff.iloc[:, 2:].copy()

# Delete columns that don't contain "Repay_perc"
columns_to_keep = [col for col in historical_repayment_percentages.columns if 'Repay_perc' in col]
historical_repayment_percentages = historical_repayment_percentages[columns_to_keep]

# Extract "Column1" and "Origination Amount" columns from dff
columns_to_add = ['Column1', 'Origination Amount']
additional_columns = dff[columns_to_add]

# Concatenate additional columns with historical_repayment_percentages
historical_repayment_percentages = pd.concat([additional_columns, historical_repayment_percentages], axis=1)

historical_repayment_percentages.columns = historical_repayment_percentages.columns.str.replace('_Repay_perc', '')
historical_repayment_percentages.columns = historical_repayment_percentages.columns.str.replace('Repayment', 'Historical_repay_percent')

historical_repayment_percentages.head()
```

Compute the expected repayment percentages for all vintages over the lifetime of the loans. Details on how the expected repayment percentages are to be computed can be found in the attached assumptions PDF below.

```python
# Initialize a list to store expected repayment percentages for each vintage
expected_repayment_percentages = []

# Historical repayment percentages columns
repayment_cols = historical_repayment_percentages.columns[2:]

# Iterate over each vintage
for index, row in historical_repayment_percentages.iterrows():
    vintage = row['Column1']
    origination_amount = row['Origination Amount']
    expected_repayments = []
    
    # Iterate over each repayment period for the vintage
    for i, col in enumerate(repayment_cols, start=1):
        repayment_percentage = row[col]
        
        # Calculate expected repayment percentage based on the formula
        if i == 1:
            expected_repayment = repayment_percentage
        elif i == 2:
            if index == 4: # Check if vintage is December 2020
                expected_repayment = 2 * expected_repayments[0]
            else:
                expected_repayment = expected_repayments[0]
        else:
            expected_repayment = repayment_percentage

        expected_repayments.append(expected_repayment)
    
    expected_repayment_percentages.append(expected_repayments)

# Convert the list of expected repayment percentages into a DataFrame
expected_repayments_df = pd.DataFrame(expected_repayment_percentages, columns=repayment_cols)

# Insert 'Column1' and 'Origination Amount' columns at the beginning of the DataFrame
expected_repayments_df.insert(0, 'Column1', historical_repayment_percentages['Column1'])
expected_repayments_df.insert(1, 'Origination Amount', historical_repayment_percentages['Origination Amount'])

# Rename columns with 'expected_repay'
expected_repayments_df.columns = expected_repayments_df.columns.str.replace('Historical_repay_percent', 'expected_repayment_perc')
expected_repayments_df.columns = expected_repayments_df.columns.str.replace('_Repay_perc', '')

expected_repayments_df.head()
```

From the expected repayment percentages, compute the forecasted cash flows using the origination amounts

```python
# Set display options to show the entire number
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
# Compute forecasted cash flows
forecasted_cash_flows = expected_repayments_df.copy()

# Iterate over each vintage
for col in expected_repayments_df.columns[2:]:
    forecasted_cash_flows[col] = (forecasted_cash_flows[col] / 100) * forecasted_cash_flows['Origination Amount']


# Rename columns with 'forecasted_cash_flow'

forecasted_cash_flows.columns =forecasted_cash_flows .columns.str.replace('expected_repayment_perc', 'Forecasted_CashFlow')

forecasted_cash_flows.head()
```

Using the assumed discount rate, derive the present value of the forecasted cash flows and of the portfolio. Donâ€™t forget to convert the annual interest rate to a monthly interest rate

```python
# Define the annual discount rate
annual_discount_rate = 0.025

# Convert the annual discount rate to a monthly discount rate
monthly_discount_rate = annual_discount_rate / 12

# Calculate the number of periods (months) for forecasting
num_periods = 30

# Initialize present value list
present_values = []

# Iterate over each row in the forecasted cash flow DataFrame
for index, row in forecasted_cash_flows.iterrows():
    origination_amount = row['Origination Amount']
    forecasted_cash_flows_row = row.iloc[2:]  # Get forecasted cash flows for the row

    # Initialize the present value for the row
    present_value_row = 0

    # Iterate over each forecasted cash flow
    for i, cash_flow in enumerate(forecasted_cash_flows_row, start=1):
        # Calculate the present value of the cash flow
        present_value_cash_flow = cash_flow / ((1 + monthly_discount_rate) ** i)
        present_value_row += present_value_cash_flow  # Add present value of cash flow to row

    present_values.append(present_value_row)  # Add present value of row to list

# Add present values to DataFrame
forecasted_cash_flows['Present Value'] = present_values

# Calculate the present value of the portfolio
present_value_portfolio = sum(present_values)

print("Present Value of the Forecasted Cash Flows:")
forecasted_cash_flows[['Column1', 'Origination Amount', 'Present Value']].round(2)
print("\nPresent Value of the Portfolio:")
print(round(present_value_portfolio,2))
```
