To address this imbalance, you can apply resampling techniques such as:

Over-sampling: Increase the number of instances in the minority class (subscribed) by generating synthetic samples or replicating existing samples. Techniques like SMOTE (Synthetic Minority Over-sampling Technique) can be used for this purpose.

```python
from imblearn.over_sampling import SMOTE

# Instantiate the SMOTE
smote = SMOTE(random_state=42)

# fit predictor and target variable
X_smote, y_smote = smote.fit_resample(X_train, y_train)

print('Original dataset shape', len(X_train))
print('Resampled dataset shape', len(X_smote))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc


# Split data into train and test sets

# Define models
models = {
    "Logistic Regression": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(probability=True),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Naive Bayes": GaussianNB(),
    "ANN": MLPClassifier()
}

# Train and evaluate models
results = {}
roc_auc_scores = {}
feature_importances = {}
for name, model in models.items():
    model.fit(X_smote, y_smote)
    y_pred = model.predict(X_smote)
    accuracy = accuracy_score(y_smote, y_pred)
    classification_rep = classification_report(y_smote, y_pred)
    conf_matrix = confusion_matrix(y_smote, y_pred)
    
    # Compute ROC AUC curve
    fpr, tpr, thresholds = roc_curve(y_smote, model.predict_proba(X_smote)[:,1])
    roc_auc = auc(fpr, tpr)
    roc_auc_scores[name] = roc_auc
    
    # Compute feature importance
    if hasattr(model, 'feature_importances_'):
        feature_importances[name] = model.feature_importances_
    elif hasattr(model, 'coef_'):
        feature_importances[name] = np.abs(model.coef_)

    results[name] = {
        "accuracy": accuracy,
        "classification_report": classification_rep,
        "confusion_matrix": conf_matrix
    }

    # Plot ROC AUC curve
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {name}')
    plt.legend(loc="lower right")
    plt.show()

    # Plot feature importance with feature names labeled
    plt.figure()
    if hasattr(model, 'feature_importances_'):
        # Plot feature importance with feature names
        feature_names = ['job', 'marital', 'education', 'default', 'housing', 'loan','age_group']

        plt.bar(feature_names, feature_importances[name])
        plt.xlabel('Feature')
        plt.ylabel('Importance')
        plt.title(f'Feature Importance - {name}')
        plt.xticks(rotation=45)
    elif hasattr(model, 'coef_'):
        # Plot feature importance with feature names
        feature_names = ['job', 'marital', 'education', 'default', 'housing', 'loan','age_group']
        plt.bar(feature_names, feature_importances[name][0])
        plt.xlabel('Feature')
        plt.ylabel('Absolute Coefficient')
        plt.title(f'Feature Importance - {name}')
        plt.xticks(rotation=45)
    
    plt.show()

# Display results
for name, metrics in results.items():
    print(f"Model: {name}")
    print(f"Accuracy: {metrics['accuracy']}")
    print("Classification Report:")
    print(metrics['classification_report'])
    print("Confusion Matrix:")
    print(metrics['confusion_matrix'])
    print("\n")

# Display feature importance
for name, importance in feature_importances.items():
    print(f"Feature Importance - {name}:")
    print(importance)
    print("\n")
```
